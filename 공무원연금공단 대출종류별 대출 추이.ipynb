```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

plt.rc('font', family='NanumBarunGothic')

import warnings
warnings.filterwarnings('ignore')
```

```python
data = pd.read_csv('./공무원연금공단 대출종류별 대출 추이.csv',encoding='cp949')
```

```python
data.head()
```

```python
data.info()
```

```python
def make_plot(df,col):
  plot_df1 = df[df['구분'] == '건수']
  plt.figure(figsize=(20,10))
  plt.subplot(1,2,1)
  sns.barplot(x=plot_df1['연도'],y=plot_df1[col])
  plt.xticks(rotation=45)
  plt.title(f'년도별 {col} 건수',size=20)

  # 단위 : 백만원
  plot_df2 = df[df['구분'] == '금액']
  plt.subplot(1,2,2)
  sns.barplot(x=plot_df2['연도'],y=plot_df2[col])
  plt.xticks(rotation=45)
  plt.title(f'년도별 {col} 금액',size=20)
  
  plt.show()
```

```python
for col in data.columns[2:]:
  make_plot(data,col)
```
