```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

plt.rc('font', family='NanumBarunGothic')

import warnings
warnings.filterwarnings('ignore')
```

```python
data = pd.read_csv('./경기도 수원시 일평균 유동인구 데이터.csv',encoding='cp949')
```

```python
data.head()
```

```python
data['격자 id 번호'] = data['격자 id 번호'].astype(str)
```

```python
data.info()
```

```python
# 일평균 유동인구 boxplot
plt.figure(figsize=(12,7))
sns.boxplot(y=data['일평균 유동인구'])
plt.show()
```

```python
# 유동인구 top10 격자
plt.figure(figsize=(20,10))
plt.subplot(1,2,1)
sns.boxplot(y=data['일평균 유동인구'])
plt.title()

plt.subplot(1,2,2)
plot_df = data.sort_values('일평균 유동인구')[::-1]
sns.barplot(x=plot_df['격자 id 번호'][:10], y=plot_df['일평균 유동인구'][:10])
plt.show()
```

```python
def make_plot(df,col):
  plt.figure(figsize=(20,10))
  plt.subplot(1,2,1)
  sns.boxplot(y=df[col])
  plt.title(f'{col} boxplot',size=20)

  plt.subplot(1,2,2)
  plot_df = df.sort_values(col)[::-1]
  sns.barplot(x=plot_df['격자 id 번호'][:10], y=plot_df[col][:10])
  plt.title(f'{col} top 10',size=20)
  
  plt.show()
```

```python
make_plot(data,'일평균 유동인구')
```

```python
make_plot(data,'일평균 남성 유동인구')  
```
