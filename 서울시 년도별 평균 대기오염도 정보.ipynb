```python
data = pd.read_csv('./서울시 년도별 평균 대기오염도 정보.csv',encoding='cp949')
```

```python
data = data.fillna(0)
```

```python
data.info()
```

```python
def count_plot(df,col_name):
  print(df[col_name].value_counts()[:10])
  plt.figure(figsize=(12,7))
  sns.countplot(df[col_name])
  plt.xticks(rotation=45)
  plt.show()
```

```python
count_plot(data,'측정년도')
```

```python
plt.figure(figsize=(12,7))
sns.barplot(x = data['측정소명'].value_counts().index,y = data['측정소명'].value_counts())
plt.xticks(rotation=90)
plt.show()
```

```python
plt.figure(figsize=(30,10))
plt.subplots_adjust(hspace=0.5)
for i in range(6):
  plt.subplot(3,2,i+1)
  sns.distplot(data.iloc[:,2+i])
  plt.xlabel('')
  plt.ylabel('')
  plt.title(data.columns[2+i])
plt.show()
```

```python
year_df = data.groupby('측정년도').mean()

plt.figure(figsize=(30,10))
plt.subplots_adjust(hspace=0.5)
for i in range(6):
  plt.subplot(3,2,i+1)
  sns.barplot(x = year_df.index , y = year_df.iloc[:,i])
  plt.xticks(rotation=45)
  plt.xlabel('')
  plt.ylabel('')
  plt.title(year_df.columns[i])
plt.show()
```

```python
year_df = data.groupby('측정소명').mean()

plt.figure(figsize=(30,10))
plt.subplots_adjust(hspace=0.5)
for i in range(6):
  plt.subplot(3,2,i+1)
  sns.barplot(x = year_df.iloc[:,1].sort_values()[::-1].index , y = year_df.iloc[:,1].sort_values()[::-1])
  plt.xticks(rotation=90)
  plt.xlabel('')
  plt.ylabel('')
  plt.title(year_df.columns[1+i])
plt.show()
```

```python
corr = data.iloc[:,2:].corr()

plt.figure(figsize=(15,7))
sns.heatmap(corr,
            annot = True, 
            linewidths = 0.4, 
            linecolor = 'white', 
            fmt = '.1f', 
            cmap = plt.cm.PuBu) 

plt.show()
```
